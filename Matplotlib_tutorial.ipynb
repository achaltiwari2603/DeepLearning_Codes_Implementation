```python
import matplotlib as mlp
import matplotlib.pyplot as plt
```

We will use the plt.style directive to choose appropriate aesthetic styles for our figures. Here we will set the classic style, which ensures that the plots we create use the classic Matplotlib style:

```python
plt.style.use('classic')
```

```python
import numpy as np
x=np.linspace(0,10,100)
fig=plt.figure()
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
```

```python
fig.canvas.get_supported_filetypes()
```

Simple Line Plots

```python
plt.style.use('seaborn-whitegrid')
fig=plt.figure()
axe=plt.axes()
x=np.linspace(0,10)
axe.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x-1),'-',label='sin(x-1)')
plt.plot(x,np.sin(x-2),'--',label='sin(x-2)')
plt.xlim(0,10)
plt.ylim(-5,5)
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

Scatter Plots

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
```
